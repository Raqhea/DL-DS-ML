```
# I'm trying to apply a simple kernel to random shaped input matrix. Which is a simple version of image processing
import numpy as np
import pandas as pd
```

```
maxRow = 8
maxCol = 8
#
R_in, C_in = round(np.random.random() * maxRow) + 2, round(np.random.random() * maxCol) + 2
random_matrix = np.random.random((R_in,C_in)) * 100
print(R_in,C_in)
print(random_matrix.shape)
pd.DataFrame(random_matrix)
```

```
# Now we can set a simple kernel of size 2x2
R_kernel, C_kernel = 2, 2
kernel = np.array([[1,0],[0,2]])
kernel
```

```
# Find the size of output matrix
R_out, C_out = R_in - 2 + 1, C_in - 2 + 1
# Create the output matrix
output = np.zeros((R_out, C_out))
output
```

```
# Now we can do the processing stuff
for i in range(0, R_out):
  for j in range(0, C_out):
    for ii in range(0, R_kernel):
      for jj in range(0, C_kernel):
        output[i,j] += random_matrix[i+ii,j+jj] * kernel[ii,jj]
pd.DataFrame(output)
```
