```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Start with xor dataset
def generate_xor(N):
  N = 1000
  X = np.random.random((N,2)) * 2 - 1
  Y = np.zeros(N)
  Y[(X[:, 0] < 0) & (X[:, 1] > 0) ] = 1
  Y[(X[:, 0] > 0) & (X[:, 1] < 0) ] = 1
  return X,Y
```

```
X,Y = generate_xor(1000)
plt.scatter(X[:, 0], X[:, 1], c = Y)
```

```
# continue with donut dataset
def generate_donut(R_inner, R_outer, N):
  R1 = np.random.randn(N//2) + R_inner
  R2 = np.random.randn(N//2) + R_outer
  theta_r1 = np.pi * 2 * np.random.random(N//2)
  theta_r2 = np.pi * 2 * np.random.random(N//2)

  X_inner = np.concatenate([ [ R1 * np.cos(theta_r1) ], [ R1 * np.sin(theta_r1) ] ]).T
  X_outer = np.concatenate([ [ R2 * np.cos(theta_r2) ], [ R2 * np.sin(theta_r2) ] ]).T

  X = np.concatenate((X_inner,X_outer))
  Y = np.array( ([0] * (N//2)) + ([1] * (N//2)) ) 

  return X,Y
```

```
X, Y = generate_donut(5,15, 1000)
```

```
plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = 'RdBu')
```

```
def generate_sinwave(N, noise = True, noise_density = 1, freq = 0.2):
  assert (noise and (0 < noise_density <= 10)), "Noise level should be less or equal than 10!"
  X = np.sin(freq * np.arange(N))
  if noise:
    X += np.random.random(N) * (noise_density/10)
  return X
```

```
X = generate_sinwave(200, freq = .1, noise_density = 3)
plt.plot(X)
```

```
def generate_coswave(N, noise = True, noise_density = 1, freq = 0.2):
  assert (noise and (0 < noise_density <= 10)), "Noise level should be less or equal than 10!"
  X = np.cos(freq * np.arange(N))
  if noise:
    X += np.random.random(N) * (noise_density/10)
  return X
X = generate_coswave(200, freq = .1, noise_density = 3)
plt.plot(X)
```
